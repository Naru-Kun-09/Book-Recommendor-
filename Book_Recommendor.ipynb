```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arashnic/book-recommendation-dataset")

print("Path to dataset files:", path)
```

```python
import os
all_files = os.listdir(path)
```

```python
import pandas as pd
import numpy as np

final_path = os.path.join(path,all_files[0])
df = pd.read_csv(final_path)
```

```python
df.drop_duplicates(subset='Book-Title',inplace=True)
```

```python
df.duplicated(subset='Book-Title').sum()
```

```python
sample_size = 15000
df = df.sample(n=sample_size,replace=False,random_state=490)
df = df.reset_index()
df = df.drop('index',axis=1)
df.head()
```

```python
def remove_spaces(author):
    result = str(author).lower()
    return (result.replace(' ',''))
```

```python
df['Book-Author'] = df['Book-Author'].apply(remove_spaces)
df.head()
```

```python
df['Book-Title'] = df['Book-Title'].str.lower()
df['Publisher'] = df['Publisher'].str.lower()
df.head()
```

```python
clean_df = df.drop(['ISBN','Year-Of-Publication','Image-URL-S', 'Image-URL-M', 'Image-URL-L'],axis=1)

clean_df['data'] = clean_df[clean_df.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),axis =1
)
print(clean_df['data'].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(clean_df['data'])
```

```python
vectorized
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)
similarities
```

```python
df = pd.DataFrame(similarities, columns=df['Book-Title'], index= df['Book-Title']).reset_index()
df.head()
```

```python
input_book = 'far beyond the stars (star trek deep space nine)'
recommendations = pd.DataFrame(df.nlargest(11,input_book)['Book-Title'])
recommendations = recommendations[recommendations['Book-Title'] != input_book]
recommendations
```
